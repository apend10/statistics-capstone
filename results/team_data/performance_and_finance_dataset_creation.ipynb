### The 'Squad' names in performance_df are different from the 'Team' names in financial_df. We need to map this

```python
# Team name mapping from performance_df (Squad) to financial_df (Team)
team_mapping = {
    'Arsenal': 'arsenal fc',
    'Bournemouth': 'afc bournemouth',
    'Brighton': 'brighton & hove albion',
    'Burnley': 'burnley fc',
    'Chelsea': 'chelsea fc',
    'Crystal Palace': 'crystal palace',
    'Everton': 'everton fc',
    'Huddersfield': 'huddersfield town',
    'Leicester City': 'leicester city',
    'Liverpool': 'liverpool fc',
    'Manchester City': 'manchester city',
    'Manchester Utd': 'manchester united',
    'Newcastle Utd': 'newcastle united',
    'Southampton': 'southampton fc',
    'Stoke City': 'stoke city',
    'Swansea City': 'swansea city',
    'Tottenham': 'tottenham hotspur',
    'Watford': 'watford fc',
    'West Brom': 'west bromwich albion',
    'West Ham': 'west ham united',
    'Cardiff City': 'cardiff city',
    'Fulham': 'fulham fc',
    'Wolves': 'wolverhampton wanderers',
    'Aston Villa': 'aston villa',
    'Norwich City': 'norwich city',
    'Sheffield Utd': 'sheffield united',
    'Leeds United': 'leeds united',
    'Brentford': 'brentford fc',
    "Nott'ham Forest": 'nottingham forest',
    'Luton Town': 'luton town',
    'Ipswich Town': 'ipswich town'
}

# Create the 'Team' column in performance_df
performance_df['Team'] = performance_df['Squad'].map(team_mapping)

# Verification checks
print("="*60)
print("TEAM NAME MAPPING VERIFICATION")
print("="*60)

# Check 1: Verify all Squad names were mapped
unmapped = performance_df[performance_df['Team'].isnull()]['Squad'].unique()
if len(unmapped) > 0:
    print(f"\n❌ WARNING: {len(unmapped)} squad names could not be mapped:")
    for squad in unmapped:
        print(f"   - {squad}")
else:
    print("\n✓ All Squad names successfully mapped to Team names")

# Check 2: Show sample of mapping
print(f"\nSample of Squad → Team mapping:")
sample_mapping = performance_df[['Squad', 'Team']].drop_duplicates().head(10)
print(sample_mapping.to_string(index=False))

# Check 3: Verify Team names exist in financial_df
financial_teams = set(financial_df['Team'].unique())
performance_teams = set(performance_df['Team'].dropna().unique())

teams_not_in_financial = performance_teams - financial_teams
if len(teams_not_in_financial) > 0:
    print(f"\n⚠ WARNING: {len(teams_not_in_financial)} teams in performance_df not found in financial_df:")
    for team in teams_not_in_financial:
        print(f"   - {team}")
else:
    print(f"\n✓ All {len(performance_teams)} mapped teams exist in financial_df")

# Check 4: Show counts
print(f"\nTeam counts:")
print(f"  - Unique Squad names in performance_df: {performance_df['Squad'].nunique()}")
print(f"  - Unique Team names mapped: {performance_df['Team'].nunique()}")
print(f"  - Unique Team names in financial_df: {financial_df['Team'].nunique()}")

# Check 5: Show the final structure
print(f"\n✓ Performance dataframe now has {len(performance_df.columns)} columns")
print(f"✓ 'Team' column added at position {list(performance_df.columns).index('Team')}")

# Display first few rows showing Year, Squad, and Team columns
print("\n" + "="*60)
print("SAMPLE DATA (showing Year, Squad, Team)")
print("="*60)
print(performance_df[['Year', 'Squad', 'Team']].head(10).to_string(index=False))
```

### Combining both the dfs!

```python

# Store original counts
print(f"\nOriginal dataframe sizes:")
print(f"  - performance_df: {len(performance_df)} rows, {len(performance_df.columns)} columns")
print(f"  - financial_df: {len(financial_df)} rows, {len(financial_df.columns)} columns")

# Check unique Year-Team combinations before merge
perf_combos = set(zip(performance_df['Year'], performance_df['Team']))
fin_combos = set(zip(financial_df['Year'], financial_df['Team']))

print(f"\nUnique (Year, Team) combinations:")
print(f"  - performance_df: {len(perf_combos)}")
print(f"  - financial_df: {len(fin_combos)}")
print(f"  - Overlap: {len(perf_combos & fin_combos)}")

# Merge the dataframes
final_df = performance_df.merge(
    financial_df,
    on=['Year', 'Team'],
    how='inner',  # Using inner join since we only want rows that exist in both
    suffixes=('_perf', '_fin')
)

print(f"\n✓ Dataframes merged successfully!")
print(f"  - final_df: {len(final_df)} rows, {len(final_df.columns)} columns")
```

```python
# Verification checks
print("\n" + "="*60)
print("DATA INTEGRITY CHECKS")
print("="*60)

# Check 1: Verify no duplicates
duplicates = final_df.duplicated(subset=['Year', 'Team']).sum()
print(f"\nCheck 1 - Duplicate (Year, Team) pairs: {duplicates}")
if duplicates > 0:
    print("WARNING: Found duplicate entries!")

# Check 2: Check for missing data
null_counts = final_df.isnull().sum()
cols_with_nulls = null_counts[null_counts > 0]
if len(cols_with_nulls) > 0:
    print(f"\nCheck 2 - Missing data:")
    print(f"  Columns with missing values: {len(cols_with_nulls)}")
    print(f"  Total missing values: {null_counts.sum()}")
    print(f"  Details:")
    for col, count in cols_with_nulls.items():
        print(f"    - {col}: {count} missing values")
else:
    print(f"\nCheck 2 - ✓ No missing values detected!")

# Check 3: Year range
print(f"\nCheck 3 - Year range:")
print(f"  - Min year: {final_df['Year'].min()}")
print(f"  - Max year: {final_df['Year'].max()}")
print(f"  - Unique years: {sorted(final_df['Year'].unique())}")

# Check 4: Teams included
print(f"\nCheck 4 - Teams:")
print(f"  - Unique teams: {final_df['Team'].nunique()}")
print(f"  - Teams: {sorted(final_df['Team'].unique())}")

# Check 5: Key columns verification
print(f"\nCheck 5 - Key columns present:")
key_cols = ['Year', 'Team', 'Squad', 'Points', 'Position', 
            'Performance_Gls', 'Average_Market_Value_Normalized_by_Avg']
for col in key_cols:
    present = "✓" if col in final_df.columns else "✗"
    print(f"  {present} {col}")

# Save the combined dataframe
output_file = 'final_combined_squad_stats.csv'
final_df.to_csv(output_file, index=False)
print(f"\n✓ Final dataframe saved to: {output_file}")
```
